## 10. Simulación de partidas y ranking en shooter online
Simule varias partidas de un juego de disparos y genere un ranking basado en el puntaje
total de cada jugador. Se utilizará el siguiente sistema de puntuación:

```python
import sys
import os

sys.path.append(os.path.abspath("../")) 

#defino el sistema de puntuacion
pt_kill=3
pt_asistencia = 1
pt_muerte = -1

#defino la estructura para el conteo del puntaje asignando a cada jugador el puntaje inicial en 0,


score_players = {
'Shadow': {'kills': 0, 'assists': 0, 'deaths': 0 , 'MPV':0 , 'pt_final':0},
'Blaze': {'kills': 0, 'assists': 0, 'deaths': 0, 'MPV':0 , 'pt_final':0},
'Viper': {'kills': 0, 'assists': 0, 'deaths': 0, 'MPV':0 , 'pt_final':0},
'Frost': {'kills': 0, 'assists': 0, 'deaths': 0, 'MPV':0 , 'pt_final':0},
'Reaper': {'kills': 0, 'assists': 0, 'deaths': 0, 'MPV':0 , 'pt_final':0},
}


rounds = [
{
'Shadow': {'kills': 2, 'assists': 1, 'deaths': True},
'Blaze': {'kills': 1, 'assists': 0, 'deaths': False},
'Viper': {'kills': 1, 'assists': 2, 'deaths': True},
'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
},
{
'Shadow': {'kills': 0, 'assists': 2, 'deaths': False},
'Blaze': {'kills': 2, 'assists': 0, 'deaths': True},
'Viper': {'kills': 1, 'assists': 1, 'deaths': False},
'Frost': {'kills': 2, 'assists': 1, 'deaths': True},
'Reaper': {'kills': 0, 'assists': 1, 'deaths': False}
},
{
'Shadow': {'kills': 1, 'assists': 0, 'deaths': False},
'Blaze': {'kills': 2, 'assists': 2, 'deaths': True},
'Viper': {'kills': 1, 'assists': 1, 'deaths': True},
'Frost': {'kills': 0, 'assists': 1, 'deaths': False},
'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
},
{
'Shadow': {'kills': 2, 'assists': 1, 'deaths': False},
'Blaze': {'kills': 1, 'assists': 0, 'deaths': True},
'Viper': {'kills': 0, 'assists': 2, 'deaths': False},
'Frost': {'kills': 1, 'assists': 1, 'deaths': True},
'Reaper': {'kills': 1, 'assists': 1, 'deaths': False}
},
{
'Shadow': {'kills': 1, 'assists': 2, 'deaths': True},
'Blaze': {'kills': 0, 'assists': 1, 'deaths': False},
'Viper': {'kills': 2, 'assists': 0, 'deaths': True},
'Frost': {'kills': 1, 'assists': 1, 'deaths': False},
'Reaper': {'kills': 1, 'assists': 1, 'deaths': True}
}
]

round_number =0

for round in rounds:
    #defino la variable para contabilizar cada ronda
    round_number+=1
    print( f"Ranking ronda {round_number}:\n")
    this_round = {"Shadow":0,"Blaze":0 ,"Viper":0,"Frost":0,"Reaper":0}
    for player in round:
        # tomo el puntaje de cada jugador para la ronda actual
        score_ind = round[player]

        #calculo el puntaje total de cada jugados en este ronda

        this_round[player]= score_ind['kills'] *pt_kill + score_ind['assists']*pt_asistencia + score_ind['deaths']* pt_muerte

        # calculo los totales

        score_players[player]['kills'] += score_ind['kills']
        score_players[player]['assists'] += score_ind['assists']
        score_players[player]['deaths'] += score_ind['deaths']

    
    
        #sumo al puntaje final lo obtenido en esta ronda

        score_players[player]['pt_final'] +=  this_round[player]

     

    ## defino el ganador de la ronda
    winner = max(this_round, key=this_round.get)
    print (winner)
    score_players[winner]['MPV']+=1
    #se muestra el resultado final

#muestro resultado final
print(score_players)
    
```
